```python
import pandas as pd 
```

```python
CpEfvEf = pd.read_csv("LFC_DE_CpEf_hypoxic_v_Ef_hypoxic_results_ashr.csv") 
CaEfvEf = pd.read_csv("LFC_DE_CaEf_hypoxic_v_Ef_hypoxic_results_ashr.csv") 
efinfo = pd.read_csv("~/SeqFiles/OG1RF_genename_function.csv")
efinfo = efinfo.drop(['Transcript'], axis=1)
efinfo = efinfo.rename({'Gene': 'gene'}, axis=1)
efinfo
```

```python
cpef2 = CpEfvEf.rename({'log2FoldChange': 'CpEf_v_Ef_Log2FC', 'Unnamed: 0': 'gene'}, axis=1)
caef2 = CaEfvEf.rename({'log2FoldChange': 'CaEf_v_Ef_Log2FC', 'Unnamed: 0': 'gene'}, axis=1)  # new methodcp
cpef3 = cpef2[['gene','CpEf_v_Ef_Log2FC']]
caef3 = caef2[['gene','CaEf_v_Ef_Log2FC']]
# cpef3 = cpef2.drop(['B', 'C'], axis=1)
Effiles = [caef3,cpef3]
Effiles
```

```python
ef = caef3.merge(cpef3, on='gene',how='outer')
efinfo2 = efinfo.merge(ef, on='gene', how='right')
efinfo3 = efinfo2[(efinfo2["CaEf_v_Ef_Log2FC"] >= 1.5) | (efinfo2["CaEf_v_Ef_Log2FC"] <= -1.5) | (efinfo2["CpEf_v_Ef_Log2FC"] >= 1.5) | (efinfo2["CpEf_v_Ef_Log2FC"] <= -1.5)]
efinfo3
```

```python
ef.to_excel('CaCpEf4_Efresponse_master.xlsx', header=True)
efinfo2.to_excel('CaCpEf4_Efresponse_master_description.xlsx', header=True)
efinfo3.to_excel('CaCpEf4_Efresponse_master_description_LFCfilter.xlsx', header=True)
```

```python
Efsigup = efinfo2[(efinfo2['CaEf_v_Ef_Log2FC']>=2) & (efinfo2['CpEf_v_Ef_Log2FC']>=2)]
Efsigup.to_excel('CaCpEf4_Efresponse_genesglobalup.xlsx', header=True)
Efsigup.count()
```

```python
Efsigdown= efinfo2[(efinfo2['CaEf_v_Ef_Log2FC']<=-2) & (efinfo2['CpEf_v_Ef_Log2FC']<=-2)]
Efsigdown.to_excel('CaCpEf4_Efresponse_genesglobaldown.xlsx', header=True)
Efsigdown.count()
```

```python
CpEfvCp = pd.read_csv("LFC_DE_CpEf_hypoxic_v_Cp_hypoxic_results_ashr.csv") 
Cpinfo = pd.read_csv('~/SeqFiles/C_parapsilosis_CDC317_current_chromosomal_feature3.csv')
CaCpconvert = pd.read_csv('~/SeqFiles/C_albicans_SC5314_C_parapsilosis_CDC317_orthologs.csv')
CaCp2 = CaCpconvert[['C_albicans','C_par']]
Cpinfo
```

```python
Cpinfo2 = Cpinfo.merge(CaCp2, how='left', left_on='Transcript', right_on='C_par' )
Cpinfo2 = Cpinfo2.drop(['C_par'], axis=1)
Cpinfo2 = Cpinfo2.rename({'C_albicans': 'C_albicans_gene', 'name': 'gene'}, axis=1)
Cpinfo2
CpEfvCp
```

```python
CpEfvCp2 = CpEfvCp.rename({'Unnamed: 0': 'gene','log2FoldChange':'CpEf_v_Cp_Log2FC' }, axis=1)
CpEfvCp2 = CpEfvCp2[['gene','CpEf_v_Cp_Log2FC']]
CpEfvCp2
Cpinfo3 = Cpinfo2.merge(CpEfvCp2, on='gene', how='right')
Cpinfo3
```

```python
Cpinfofilter = Cpinfo3[(Cpinfo3["CpEf_v_Cp_Log2FC"] >= 1.5) | (Cpinfo3["CpEf_v_Cp_Log2FC"] <= -1.5)]
Cpinfofilter
```

```python
Cpinfo3.to_excel('CpEf4_Cpresponse_master_description.xlsx', header=True)
Cpinfofilter.to_excel('CpEf4_Cpresponse_master_description_LFCfiltered.xlsx', header=True)
```
